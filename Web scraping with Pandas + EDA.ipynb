In this project, I scraped data from the NBA players statistics website, and performed some exploratory data analysis to uncover some information from the tables. 

I used the figures for the 2021/2022 basketball year. 

```python
import pandas as pd
```

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2022_per_game.html'
```

```python
html = pd.read_html(url, header = 0)
```

```python
data2022 = html[0]
```
